```python
# Clase autovil, atributos, marca, placa, precio
# funciones leerdatos(), mostrardatos()
# crear 3 obhetos carro 1, carro 2, carro 3

class Automovil:
  def __init__(self, marca, placa, precio):
    self.marca = marca
    self.placa = placa
    self.precio = precio

  def leerdatos(self):
    self.marca = input("Ingrese la marca: ")
    self.placa = input("Ingrese la placa: ")
    self.precio = input("Ingrese el precio: ")

  def mostrardatos(self):
    print(f"Marca",self.marca)
    print(f"Placa", self.placa)
    print(f"Preci", self.precio)
    return self.marca, self.placa, self.precio

# fin de  la clase automoil
# crear 3 obhetos carro 1, carro 2, carro 3
print("carro numer 1")
carro1 = Automovil("","","")
carro1 = Automovil.leerdatos()
carro1 = Automovil.mostrardatos()
print("")
print("carro numer 2")
carro2 = Automovil("","","")
carro2 = Automovil.leerdatos()
carro2 = Automovil.mostrardatos()
print("------------")
print("carro numer 3")
carro3 = Automovil("","","")
carro3 = Automovil.leerdatos()
carro3 = Automovil.mostrardatos()
print("programa realizado por Christian Garcia NL = 21")
```
